### SP500 History

```python
sp500_csv = pd.read_csv("Resources/sp500_history.csv")
sp500_data = pd.read_csv("Resources/sp500_history.csv")
sp500_data.head()
```

```python
#check data type
sp500_data.dtypes
```

```python
#fix data type
sp500_data["Close"] = sp500_data["Close"].str.replace("$", "")
sp500_data["Close"] = sp500_data["Close"].str.replace(",", "")
```

```python
sp500_data.dtypes
```

```python
sp500_data["Close"] = sp500_data["Close"].astype('float')
print(sp500_data.dtypes)
sp500_data.head()
```

```python
#set the index
sp500_data.set_index(pd.to_datetime(sp500_data['Date'], infer_datetime_format=True), inplace=True)
sp500_data.head()
```

```python
#drop the date column
sp500_data.drop(columns = ['Date'], inplace = True)
sp500_data.head()
```

```python
#calculate the daily returns
sp500_data = sp500_data.pct_change()
sp500_data.head()
```

```python
#drop the nulls 
sp500_data = sp500_data.dropna()
sp500_data
```

```python
#change the 'Close' column
columns = ['SP500 Close']
sp500_data.columns = columns
sp500_data.head()
```

### Calculate and Plot Beta 

```python
#calculate the daily return of a single portfolio to calculate the covariance 
daily_returns_algo = algo_data.pct_change()
daily_returns_algo
```

```python
#calculate the covariance of the algo portfolio
covariance_algo = daily_returns_algo.cov()
covariance_algo
```

```python
#calculate the variance 
variance_algo = daily_returns_algo.var()
variance_algo
```

```python
beta_algo = covariance_algo / variance_algo
beta_algo
```
